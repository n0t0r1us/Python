```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
def image_translate(image,shiftX,shiftY):
   M = np.float32([[1, 0, shiftX], [0, 1, shiftY]])
   return cv2.warpAffine(image,M,(image.shape[1],image.shape[0]))
img=cv2.imread('D:/trex.png')
fig=plt.figure()
ax1,ax2=fig.subplots(1,2)
ax1.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
ax1.set_title("orignal")
pixel_down=20
pixel_right=10
# shift image along x,y axis.
shifted = image_translate(img,10,20)
ax2.imshow(shifted)
ax2.set_title("Translated image")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math
def image_rotation(image,degree):
  theta=-degree*math.pi/180
  (h,w)=image.shape[:2]
  MT1=np.float32([[1, 0, w//2], [0, 1, h//2]])
  MR=M=np.float32([[math.cos(theta), -math.sin(theta), 0],[math.sin(theta),
                                                           math.cos(theta), 0],[0,0,1]])
  MT2=np.float32([[1, 0, -w//2], [0, 1, -h//2],[0,0,1]])
  #M: final transformation matrix
  M=MT1@MR@MT2
  return cv2.warpAffine(image, M,(w,h))
img=cv2.imread('D:/trex.png')
fig=plt.figure()
ax1,ax2=fig.subplots(1,2)
ax1.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
ax1.set_title("orignal")
rotated=image_rotation(img,20)
ax2.imshow(rotated)
ax2.set_title("Rotated image")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def image_scaling(image, scaleX=1, scaleY=1):
    (h, w) = image.shape[:2]
    new_w = int(w * scaleX)
    new_h = int(h * scaleY)
    M = np.float32([[scaleX, 0, 0], [0, scaleY, 0]])
    scaled_image = cv2.warpAffine(image, M, (new_w, new_h))
    return scaled_image, (new_w, new_h)

if __name__ == '__main__':
    img = cv2.imread('D:/trex.png')
    fig = plt.figure()
    ax1, ax2 = fig.subplots(1, 2)
    ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax1.set_title("Original")
    scaled, new_size = image_scaling(img, 0.2, 0.2)
    ax2.imshow(cv2.cvtColor(scaled, cv2.COLOR_BGR2RGB))
    ax2.set_title("Scaled image")
    plt.show()
    print("New size after scaling:", new_size)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
def image_resized(image, width=None, height=None, inter=cv2.INTER_AREA):
    dim=None
    (h,w)=image.shape[:2]
    if width is None and height is None:
        return image
    if height is None:
        r = width / float(w)
        dim=(width,int(h*r))
    else:
        r = height / float(h)
        dim=(int(w*r),height)
    resize=cv2.resize(image,dim,interpolation=inter)
    return resize
if __name__=='__main__':
 img=cv2.imread('D:/trex.png')
 fig=plt.figure()
 ax1,ax2=fig.subplots(1,2)
 ax1.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
 ax1.set_title("orignal")
 resized=image_resized(img,width=50)
 ax2.imshow(resized)
 ax2.set_title("Resize image with INTER_AREA method")
 plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
plt.rcParams.update(plt.rcParamsDefault)
def image_cropping(image, start_x, end_x, start_y, end_y):
    cropped_image = image[start_y:end_y, start_x:end_x]
    return cropped_image
if __name__ == '__main__':
    img = cv2.imread('D:/trex.png')
    fig = plt.figure()
    ax1, ax2 = fig.subplots(1, 2)
    ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax1.set_title("Original")
    start_x, end_x, start_y, end_y = img.shape[1] // 2, img.shape[1], 0, img.shape[0] // 2
    cropped = image_cropping(img, start_x, end_x, start_y, end_y)
    ax2.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
    ax2.set_title("Cropped Image")
    plt.show()
```
