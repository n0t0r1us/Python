```python
#TH01
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
img=cv2.imread('D:/beach.png')
print(f"width of image: {img.shape[1]} pixels")
print(f"height of image: {img.shape[0]} pixels")
print(f"channel of image: {img.shape[2]}")
fig=plt.figure()
plt.imshow(img) #B=0, G=1, R=2
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)) #B=0, G=1, R=2
plt.show()
cv2.imwrite("du_newfigure1.png", img)
```

```python
#2.1
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
img=cv2.imread('D:/trex.png')
# ly pixel gc trn cng bn tri theo th t Blue, Green v Red
(b_tl,g_tl,r_tl)=img[0,0,:]
# ly pixel gc di cng bn phi theo th t Blue, Green v Red
(b_br,g_br,r_br)=img[-1,-1,:]
print(f"Pixels at (0, 0) - Red: {r_tl}, Green: {g_tl}, Blue: {b_tl}")
print(f"Pixels at ({img.shape[0]-1}, {img.shape[1]-1}) - Red: {r_br}, Green: {g_br}, Blue: {b_br}")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)) #B=0, G=1, R=2
plt.show()
```

```python
#2.3
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
plt.rcParams.update(plt.rcParamsDefault)
img=cv2.imread('D:/trex.png')
fig=plt.figure()
ax1,ax2=fig.subplots(1,2)
ax1.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
ax1.set_title("original")
img[0:50,0:50]=(0,0,255)
ax2.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
ax2.set_title("update pixels")
plt.show()
```

```python
#3.1
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import numpy as np
# T to ra mt hnh nn en bng numpy
canvas=np.zeros((300,300,3),dtype='uint8')
white=(255,255,255)
red=(255,0,0)
cv2.line(canvas, (0,0), (canvas.shape[0],canvas.shape[1]), white)
cv2.line(canvas, (canvas.shape[0],0), (0,canvas.shape[1]), red)
plt.imshow(canvas)
plt.xlim([0,300-1])
plt.ylim([0,300-1])
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import numpy as np
plt.rcParams.update(plt.rcParamsDefault)
canvas=np.zeros((300,300,3),dtype='uint8')
white=(255,255,255)
red=(255,0,0)
cv2.rectangle(canvas, (10,10), (canvas.shape[0]//4,canvas.shape[1]//4), white,
2)
cv2.rectangle(canvas, (100,100), (200,200), red, -1)
plt.imshow(canvas)
plt.show()
```
