##1) MÔ TẢ TẬP DỮ LIỆU
---

Tập dữ liệu thu thập được từ 303 bệnh nhân, những người có biểu hiện của chứng đau ngực, tương ứng với 303 dòng trong file dữ liệu, mỗi dòng có 14 thuộc tính cung cấp các thông tin khác nhau của một bệnh nhân. Chi tiết các thuộc tính như sau:

* **feature1**: Thuộc tính này cho biết tuổi của bệnh nhân (Loại dữ liệu: Số - Ví dụ: 65, 38... )

* **feature2**: Thộc tính này cho giới tính của bệnh nhân (Male | Female) (Loại dữ liệu: Ký tự - Ví dụ: Male)

* **feature3**: Thộc tính này cho biết loại chứng đau ngực mà bệnh nhân mắc phải, bao gồm  4 loại như bên dưới (Loại dữ liệu: ký tự - Ví dụ: Value1...)
  *  Typical angina. 
  *  Atypical angina. 
  *  Non-anginal pain. 
  *  Asymptomatic  


* **feature4**: Thộc tính này cho biết huyết áp của bệnh nhân khi nhập viện - đơn vị mmhg (Loại dữ liệu: Số)

* **feature5**: Thộc tính này cho biết lượng cholesterol trong máu - đơn vị mg/dl (Loại dữ liệu: Số)

* **feature6**: Thộc tính này cho biết đường huyết lúc đói của bệnh nhân có > 120mg/dl hay không?, True | False (Loại dữ liệu: Ký tự)

* **feature7**: Thộc tính này cho biết kết quả điện tâm đồ của bệnh nhân lúc nghỉ ngơi, có 3 giá trị  (Loại dữ liệu: số):
  * 0: Bình thường. 
  * 1: Có bất thường sóng ST-T(T-wave inversions and/or ST elevation or depression of > 0.05 mV). 
  * 2: Phì đại thất trái có thể xảy ra...

* **feature8**: Thộc tính này cho biết nhịp tim tối đa đạt được (Loại dữ liệu: Số)

* **feature9**: Thộc tính này cho biết khi bệnh nhân tập thể dục có gấy đau thắt ngực không, Yes | No (Loại dữ liệu: ký tự)

* **feature10**: Thộc tính này cho biết giá trị của ST trầm cảm gây ra bởi tập thể dục liên quan tới nghỉ ngơi (Loại dữ liệu: Số)

* **feature11**: Thộc tính này cho biết độ dốc của đỉnh đoạn STtập thể dục, Bao gồm 3 giá trị như dưới (Loại dữ liệu: Ký tự):
  * Upsloping | Us | U: Dốc lên, 
  * Flat | F: Bằng phẳng
  * Down sloping | Ds | D: Dốc xuống

* **feature12**: Thộc tính này cho biết số lượng các mạch lớn được tô bằng huỳnh quang, giá trị là các số từ 0 - 3  (Loại dữ liệu: Số - Ví dụ: 0, 3...)

* **feature13**: Thộc tính này cho biết chỉ số Thalassemia của bệnh nhân, gồm 3 giá trị số (Loại dữ liệu: Số - Ví dụ: 3, 6, 7..)
  * 3: Bình thường
  * 4: Khiếm khuyết cố định
  * 7: Khiếm khuyết có thể đảo ngược.

* **outcome**: Thuộc tính đầu ra cho biết tình trạng bệnh tim mạch của bệnh nhân này, 0: Không bị bệnh tim mạch, >0: bị bệnh tim (bệnh tim được chia ra thành các cấp độ từ 1 đến 4 )
